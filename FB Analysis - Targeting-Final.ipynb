```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font_scale=1.2,style='darkgrid')

%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
#change display into using full screen
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
#import .csv file
data = pd.read_csv("D:\Python\My Project\Data files\Fb Target analysis.csv")
data.head()
```

```python
data.shape
```

```python
#drop Data
data.drop(['Reporting Ends','Reporting Starts'],axis=1, inplace = True)
```

```python
data.drop(['Website Purchases'],axis=1, inplace = True)
```

```python

data.head()
```

```python

data.shape
```

```python
#null check
data.isnull().sum()
```

```python
data.info()
```

```python
data.head()
```

```python
#frequency bucket
data['Frequency'] = data['Frequency'].apply(lambda x:'1 to 2' if x<2
                                               else '2 to 3' if x>=2 and x<3 
                                               else '3 to 4' if x>=3 and x<4
                                               else '4 to 5' if x>=4 and x<5
                                               else 'More than 5')
```

```python
data.head()
```

```python
#split adset
data["Adset_Name"] = data['Ad Set Name']
```

```python
data.head()
```

```python
data[['dept','target']] = data.Adset_Name.str.split("-",expand=True)
```

```python
data.head()
```

```python
data.info(verbose=1)
```

```python
#Save data
data.to_csv('D:\Python\My Project\Data files\Cleaned_Data_Targeting.csv')
```

```python
#relationship visualization

plt.figure(figsize=(28, 7))


plt.subplot(1, 6, 1)
sns.barplot(x='Ad Set Name', y='Amount Spent (INR)', data=data, estimator=np.sum,ci=None)
plt.title("Total Amount Spent")
plt.xticks(rotation=90)



plt.subplot(1, 6, 2)
sns.barplot(x='Ad Set Name', y='Clicks (All)', data=data, estimator=np.sum,ci=None)
plt.title("Total Clicks")
plt.xticks(rotation=90)


plt.subplot(1, 6, 3)
sns.barplot(x='Ad Set Name', y='CTR (All)', data=data, estimator=np.mean,ci=None)
plt.title("CTR")
plt.xticks(rotation=90)


plt.subplot(1, 6, 4)
sns.barplot(x='Ad Set Name', y='Results', data=data, estimator=np.sum,ci=None)
plt.title("Total Conversions")
plt.xticks(rotation=90)


plt.subplot(1, 6, 5)

sns.barplot(x='Ad Set Name', y='Cost per Result', data=data, estimator=np.mean,ci=None)
plt.title("Avg. Cost per Conversion")
plt.xticks(rotation=90)


plt.subplot(1,6, 6)
sns.barplot(x='Ad Set Name', y='Result Rate', data=data, estimator=np.mean,ci=None)
plt.title("CVR")
plt.xticks(rotation=90)


plt.tight_layout(pad=0.7)
plt.show()
```

```python
#Target Split Data

plt.figure(figsize=(28, 7))

plt.subplot(1, 6, 1)
sns.barplot(x='target', y='Amount Spent (INR)', data=data, estimator=np.sum,ci=None)
plt.title("Target Vs Amount Spent")
plt.xticks(rotation=90)

plt.subplot(1, 6, 2)
sns.barplot(x='target', y='Clicks (All)', data=data, estimator=np.sum,ci=None)
plt.title("Target Vs Clicks")
plt.xticks(rotation=90)


plt.subplot(1, 6, 3)
sns.barplot(x='target', y='CTR (All)', data=data, estimator=np.sum,ci=None)
plt.title("Target Vs CTR")
plt.xticks(rotation=90)

plt.subplot(1, 6, 4)
sns.barplot(x='target', y='Results', data=data, estimator=np.sum,ci=None)
plt.title("Target Vs Results")
plt.xticks(rotation=90) 

plt.subplot(1, 6, 5)
sns.barplot(x='target', y='Cost per Result', data=data, estimator=np.sum,ci=None)
plt.title("Target Vs Cost per Result")
plt.xticks(rotation=90)
```

```python
#Streams Split Data

plt.figure(figsize=(28 , 6))
plt.subplot(1, 6, 1)
sns.barplot(x='dept', y='Amount Spent (INR)', data=data, estimator=np.sum,ci=None)
plt.title("Streams Vs Total Amount Spent")
plt.xticks(rotation=90)

plt.subplot(1, 6, 2)
sns.barplot(x='dept', y='Clicks (All)', data=data, estimator=np.sum,ci=None)
plt.title(" Streams Vs Total Clicks")
plt.xticks(rotation=90)

plt.subplot(1, 6, 3)
sns.barplot(x='dept', y='CTR (All)', data=data, estimator=np.sum,ci=None)
plt.title("Streams Vs CTR")
plt.xticks(rotation=90)

plt.subplot(1, 6, 4)
sns.barplot(x='dept', y='Results', data=data, estimator=np.sum,ci=None)
plt.title("Streams Vs Total Conversions")
plt.xticks(rotation=90)

plt.subplot(1, 6, 5)
sns.barplot(x='dept', y='Cost per Result', data=data, estimator=np.sum,ci=None)
plt.title("Streams Vs Avg. Cost per Conversion")
plt.xticks(rotation=90)

plt.subplot(1, 6, 6)
sns.barplot(x='dept', y='Result Rate', data=data, estimator=np.sum,ci=None)
plt.title("Streams Vs CVR")
plt.xticks(rotation=90)

plt.tight_layout(pad=0.7)
plt.show()
```

```python
#Granular Analysis

sns.catplot(y='dept',x='Amount Spent (INR)',col='target',data=data,kind='bar',aspect=1.3, estimator=np.sum,ci=None)
plt.subplots_adjust(top=0.85)
plt.suptitle('Total Amount Spent on Each Ad',fontsize=20)


sns.catplot(y='dept',x='Clicks (All)',col='target',data=data,kind='bar',aspect=1.3, estimator=np.sum,ci=None)
plt.subplots_adjust(top=0.85)
plt.suptitle('Total Clicks (All)',fontsize=20)

sns.catplot(y='dept',x='CTR (All)',col='target',data=data,kind='bar',aspect=1.3, estimator=np.sum,ci=None)
plt.subplots_adjust(top=0.85)
plt.suptitle('CTR',fontsize=20)

sns.catplot(y='dept',x='Results',col='target',data=data,kind='bar',aspect=1.3, estimator=np.sum,ci=None)
plt.subplots_adjust(top=0.85)
plt.suptitle('Total Conversions',fontsize=20)

sns.catplot(y='dept',x='Cost per Result',col='target',data=data,kind='bar',aspect=1.3, estimator=np.sum,ci=None)
plt.subplots_adjust(top=0.85)
plt.suptitle('Avg. Cost per Conversion',fontsize=20)

sns.catplot(y='dept',x='Result Rate',col='target',data=data,kind='bar',aspect=1.3, estimator=np.sum,ci=None)
plt.subplots_adjust(top=0.85)
plt.suptitle('CVR',fontsize=20)
```

### Let's Look at a Granular View

## 3. Effect of Targeting + Frequency + Budget on Engagement & Conversion

```python
data = data.sort_values(by=['Frequency']) 
```

### Generic View

```python

plt.figure(figsize=(28 , 6))


plt.subplot(1, 6, 1)
sns.barplot(x='Frequency', y='Amount Spent (INR)', data=data, estimator=np.sum,ci=None)
plt.title("Total Amount Spent")
plt.xticks(rotation=90)


# subplot 2
plt.subplot(1, 6, 2)
sns.barplot(x='Frequency', y='Clicks (All)', data=data, estimator=np.sum,ci=None)
plt.title("Total Clicks")
plt.xticks(rotation=90)

# subplot 3
plt.subplot(1, 6, 3)
sns.barplot(x='Frequency', y='CTR (All)', data=data, estimator=np.mean,ci=None)
plt.title("CTR")
plt.xticks(rotation=90)

# subplot 4
plt.subplot(1, 6, 4)
sns.barplot(x='Frequency', y='Results', data=data, estimator=np.sum,ci=None)
plt.title("Total Conversions")
plt.xticks(rotation=90)

# subplot 5
plt.subplot(1, 6, 5)
sns.barplot(x='Frequency', y='CPC (All)', data=data, estimator=np.mean,ci=None)
plt.title("Avg. Cost per Conversion")
plt.xticks(rotation=90)

# subplot 6
plt.subplot(1,6, 6)
sns.barplot(x='Frequency', y='Result Rate', data=data, estimator=np.mean,ci=None)
plt.title("CVR")
plt.xticks(rotation=90)


plt.tight_layout(pad=0.7)
plt.show()
```

### Granular View

```python
sns.catplot(x='Frequency',y='Amount Spent (INR)',col='Ad Set Name',data=data,kind='bar',aspect=1, estimator=np.sum,ci=None)
plt.subplots_adjust(top=0.85)
plt.suptitle('Total Amount Spent on Each Frequency',fontsize=20)

sns.catplot(x='Frequency',y='Clicks (All)',col='Ad Set Name',data=data,kind='bar',aspect=1, estimator=np.sum,ci=None)
plt.subplots_adjust(top=0.85)
plt.suptitle('Total Clicks on Each Frequency',fontsize=20)

sns.catplot(x='Frequency',y='CTR (All)',col='Ad Set Name',data=data,kind='bar',aspect=1, estimator=np.mean,ci=None)
plt.subplots_adjust(top=0.85)
plt.suptitle('Avg. Click Through Rate (CTR) of Each Frequency',fontsize=20)

sns.catplot(x='Frequency',y='Results',col='Ad Set Name',data=data,kind='bar',aspect=1, estimator=np.sum,ci=None)
plt.subplots_adjust(top=0.85)
plt.suptitle('Total Conversions from Each Frequency',fontsize=20)

sns.catplot(x='Frequency',y='CPC (All)',col='Ad Set Name',data=data,kind='bar',aspect=1, estimator=np.mean,ci=None)
plt.subplots_adjust(top=0.85)
plt.suptitle('Avg. Cost per Conversion of Each Frequency',fontsize=20)

sns.catplot(x='Frequency',y='Result Rate',col='Ad Set Name',data=data,kind='bar',aspect=1, estimator=np.mean,ci=None)
plt.subplots_adjust(top=0.85)
plt.suptitle('Avg. Conversion Rate of Each Frequency',fontsize=20)
```
